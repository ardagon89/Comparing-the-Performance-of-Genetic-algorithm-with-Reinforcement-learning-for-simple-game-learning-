```python
import networkx as nx
import matplotlib.pyplot as plt
    
if __name__ == '__main__':
    plt.figure(figsize=(10,8))
    
    # line 1 points 
    x1 = ['128M','256M','512M'] 
    y1 = [float(x) for x in "2.4E-4 4.0E-4 8.0E-4".split()]
    # plotting the line 1 points  
    plt.plot(x1, y1, label = "HybridRMQ") 

    # line 2 points 
    x2 = ['128M','256M','512M'] 
    y2 = [float(x) for x in "2.8E-4 4.0E-4 0.0036".split()]
    # plotting the line 2 points  
    plt.plot(x2, y2, label = "FischerHeunRMQ")     

    # setting x and y axis range 
    plt.ylim(0,0.004) 

    # naming the x axis 
    plt.xlabel('Array size - millions') 
    # naming the y axis 
    plt.ylabel('Time - milliseconds') 
    # giving a title to my graph 
    plt.title('Avg. Query Time') 

    # show a legend on the plot 
    plt.legend() 

    # function to show the plot 
    plt.show() 
```
